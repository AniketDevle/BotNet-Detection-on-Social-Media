```python
import pandas as pd
from pandas import json_normalize
import json
import numpy as np
```

```python
def get_screen_name_to_retweeted_status(retweeted_status=None,extended_tweet=None,quoted_status=None):
    if retweeted_status != '':
        return retweeted_status['user']['screen_name']

def get_screen_name_to_quoted_status(quoted_status=None):
    if quoted_status != '' :
        return quoted_status['user']['screen_name']
def get_retweet_id_quoted_status(quoted_status=None):
    if quoted_status != '':
        return quoted_status['id']
def get_retweet_id_retweeted_status(retweeted_status=None):
    if retweeted_status != '':
        return retweeted_status['id']
def get_screen_name_from(user):
    return user['screen_name']
def get_url_from_retweeted_status(retweeted_status=None):
    if retweeted_status!= '':
        extended_tweet = retweeted_status.get('extended_tweet')
        try:
            if extended_tweet:
                return retweeted_status['extended_tweet']['entities']['urls'][0]['expanded_url']
            else:
                return retweeted_status['entities']['urls'][0]['expanded_url']
        except:
            pass
            
def get_url_from_quoted_status(quoted_status=None):
    if quoted_status!= '':
        extended_tweet = quoted_status.get('extended_tweet')
        try:
            if extended_tweet:
                return quoted_status['extended_tweet']['entities']['urls'][0]['expanded_url']
            else:
                return quoted_status['entities']['urls'][0]['expanded_url']
        except:
            pass

def get_url_from_extended_tweet(extended_tweet=None):
    if extended_tweet!= '':
        try:
            return extended_tweet['entities']['urls'][0]['expanded_url']
        except:
            pass
def get_url_from_entities(entities=None):
    if entities!='':
        try:
            return entities['urls'][0]['expanded_url']
        except:
            pass
```

```python
def get_data(file_name):
    df1 = pd.read_csv(file_name, names=['tid', 'data'])
    df=df1
    df = df.join(pd.DataFrame((json.loads(d)) for d in df.data))
    df=df.filter(items=['tid','created_at','retweeted_status', 'extended_tweet','quoted_status','entities','user','text'])
    df = df.fillna('')
    df['screen_name_to_e'] = df['retweeted_status'].apply(get_screen_name_to_retweeted_status)
    df['screen_name_to'] = df['retweeted_status'].apply(get_screen_name_to_retweeted_status)
    df['screen_name_to_q'] = df['quoted_status'].apply(get_screen_name_to_quoted_status)
    df['screen_name_to']= np.where(df['screen_name_to'].isnull(), df['screen_name_to_q'], df['screen_name_to'])
    df['retweet_id'] = df['retweeted_status'].apply(get_retweet_id_retweeted_status)
    df['retweet_id_q'] = df['quoted_status'].apply(get_retweet_id_quoted_status)
    df['retweet_id']= np.where(df['retweet_id'].isnull(), df['retweet_id_q'], df['retweet_id'])
    df['screen_name_from'] = df['user'].apply(get_screen_name_from)
    df['url'] = df['retweeted_status'].apply(get_url_from_retweeted_status)
    df['url_q'] = df['quoted_status'].apply(get_url_from_quoted_status)
    df['url_e'] = df['extended_tweet'].apply(get_url_from_extended_tweet)
    df['url_n'] = df['entities'].apply(get_url_from_entities)
    df['url']= np.where(df['url'].isnull(), df['url_q'], df['url'])
    df['url']= np.where(df['url'].isnull(), df['url_e'], df['url'])
    df['url']= np.where(df['url'].isnull(), df['url_n'], df['url'])
    df=df.filter(items=['tid','created_at','text', 'retweet_id','screen_name_to','screen_name_from','url','screen_name_to_e'])
    return df
```

```python
for i in range(1,61):
    file_name = "out_30_file" + str(i) + ".csv"
    data = get_data(file_name)
    data.to_csv('my_csv1.csv', mode='a', header=False)
```

```python
# code to split files refer stackoverflow
import csv
# your input file (10GB)
in_csvfile = open('CORRUPT.tweets_dump.csv', "r", errors="ignore")
# reader, that would read file for you line-by-line
reader = csv.DictReader(in_csvfile)
# number of current line read
num = 0
# number of output file
output_file_num = 1
# your output file
out_csvfile = open('output_data_{}.csv'.format(output_file_num), "w")
# writer should be constructed in a read loop,
# because we need csv headers to be already available
# to construct writer object
writer = None
for row in reader:
    num += 1
    # Here you have your data line in a row variable
    # If writer doesn't exists, create one
    if writer is None:
        writer = csv.DictWriter(
            out_csvfile,
            fieldnames=row.keys(),
            delimiter=",", quotechar='"', escapechar='"',
            lineterminator='\n', quoting=csv.QUOTE_NONNUMERIC
        )
    # Write a row into a writer (out_csvfile, remember?)
    writer.writerow(row)
    # If we got a 10000 rows read, save current out file
    # and create a new one
    if num > 30000:
        output_file_num += 1
        out_csvfile.close()
        writer = None
        # create new file
        out_csvfile = open('out_30_file{}.csv'.format(output_file_num), "w")
        # reset counter
        num = 0
# Closing the files
in_csvfile.close()
out_csvfile.close()
```
